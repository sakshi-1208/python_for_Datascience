```python
#WAF to print the length of a list(list is the parameter)

cities = ["noida", "pune", "bangalore", "kolkata"]
heros = ["spiderman", "thor",
         "ironman", "batman", "shaktiman"]

def print_len(list):
  print(len(list))

print_len(cities)
print_len(heros)
```

```python
#WAF to print the elements of a list in a single line(list is the parameter)
def elemnts(list):
  for el in list:
    print(el,end=",")

elemnts(cities)
print()
elemnts(heros)
```

```python
#WAF to find the factorial of n(n is the parameter)

num=int(input("enter the number"))
def fact(n):
  prod=1
  for i in range(1,n+1):
    prod*=i
  return prod

print(fact(num))
```

```python
def show(n):
  if(n==0):
    return
  print(n)
  show(n-1)

show(5) #5=n,4=n-1,3=n-2,2=n-3,1=n-4
```

```python
def fact(num):
  if(num==0 or num==1):
    return 1
  else:
    return num * fact(num-1)

fact(3)
```

```python
def fibb(num):
  if(num == 0 or num==1):
    return num
  else:
    return fibb(num-1)+fibb(num-2)

fibb(6)
```

```python
#WARF to calculate the sum of first n natural numbers.

def sum(num):
  if(num == 0):
    return 0
  return sum(num-1) + num

sum(5)
```

```python
#WARF function to print all elements in a list

fruits = [1,2.0,"a"]

def listval(list,idx=0):
  if (idx == len(list)):
    return
  print(list[idx],end=",")
  listval(list,idx+1)

listval(fruits)

```
