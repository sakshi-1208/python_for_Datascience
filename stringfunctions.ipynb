```python
print("mountains".upper())
print("Mountains".lower())
#Whitespace includes spaces, tabs, and newline characters
print(" yes ".strip()) #removes extra space from the string
print(" yes".lstrip()) #removes extra space from the left of the string
print("yes ".rstrip()) #removes extra space from the right of the string
print("FOREST".isnumeric())
print("1234".isnumeric())
print(" ".join(["This", "is", "a", "phrase", "joined", "by", "spaces"]))
print(".....".join(["This", "is", "a", "phrase", "joined", "by", "spaces"]))
print("This is another example".split())
```

```python
#format method in string

name="Sakshi"
number=len(name)*3

print("hello {}. Your Lucky Number is: {}".format(name,number))

#if we don't know the sequence or variable might switch the places
print("your lucky number is {number}, hello {name}".format(name=name, number= len(name)*5))
```

```python
price = 10
with_tax = price*1.09
print("Base Price: ${:.2f}. With Tax: {$:.2f}".format(price, with_tax))
```
